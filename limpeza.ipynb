```python
import pandas as pd
```

```python
# 1. Ler o arquivo CSV
df = pd.read_csv("award_winners.csv")

# 2. Filtrar valores da coluna "tournament_id" que terminam em 0, 2, 4, 6, 8
df_filtrado = df[df['tournament_id'].astype(str).str.endswith(('0','2','4','6','8'))]

# 3. Salvar o resultado
df_filtrado.to_csv("award_winners_filtrado.csv", index=False)

print("Arquivo filtrado salvo como 'award_winners_filtrado.csv'")
```

```python
import pandas as pd

# Carregar os dois arquivos
players = pd.read_csv("players.csv")
squads = pd.read_csv("squads_filtrado.csv")

# Filtrar players que possuem player_id presente em squads_filtrado
players_filtrado = players[players['player_id'].isin(squads['player_id'])]

# Salvar o resultado
players_filtrado.to_csv("players_filtrado.csv", index=False)

print("Arquivo 'players_filtrado.csv' criado com sucesso!")
```

```python
import pandas as pd

# Ler o arquivo
tournaments = pd.read_csv("tournaments.csv")

# Filtrar valores da coluna "year" que terminam em 0, 2, 4, 6 ou 8
tournaments_filtrado = tournaments[
    tournaments["year"].astype(str).str.endswith(('0', '2', '4', '6', '8'))
]

# Salvar o arquivo filtrado
tournaments_filtrado.to_csv("tournaments_filtrado.csv", index=False)

print("Arquivo 'tournaments_filtrado.csv' criado com sucesso!")
```
