```python
import pandas as pd
```

```python
# Step 1: Load the CSV file
df = pd.read_csv("award_winners.csv")

# Step 2: Filter tournaments related to men's World Cups
# Tournament IDs ending in even numbers represent men's competitions
df_filtered = df[df["tournament_id"].astype(str).str.endswith(("0", "2", "4", "6", "8"))]

# Step 3: Save the filtered dataset
df_filtered.to_csv("award_winners_filtered.csv", index=False)

print("Filtered file saved as 'award_winners_filtered.csv'")
```

```python
# Step 1: Load the CSV file
df = pd.read_csv("squads.csv")

# Step 2: Filter tournaments related to men's World Cups
# Tournament IDs ending in even numbers represent men's competitions
df_filtrado = df[df['tournament_id'].astype(str).str.endswith(('0','2','4','6','8'))]

# Step 3: Save the filtered dataset
df_filtrado.to_csv("squads_filtered.csv", index=False)

print("Filtered file saved as 'squads_filtered.csv'")
```

```python
# Load input datasets
players = pd.read_csv("players.csv")
squads = pd.read_csv("squads_filtered.csv")

# Filter players whose player_id is present in the filtered squads dataset
players_filtrado = players[players['player_id'].isin(squads['player_id'])]

# Step 3: Save the filtered dataset
players_filtrado.to_csv("players_filtered.csv", index=False)

print("Filtered file saved as 'players_filtered.csv'!")
```

```python
# Load input datasets
tournaments = pd.read_csv("tournaments.csv")

# Filter records where the "year" value ends with 0, 2, 4, 6, or 8
tournaments_filtrado = tournaments[
    tournaments["year"].astype(str).str.endswith(('0', '2', '4', '6', '8'))
]

# Step 3: Save the filtered dataset
tournaments_filtrado.to_csv("tournaments_filtered.csv", index=False)

print("Filtered file saved as 'tournaments_filtered.csv'!")
```
